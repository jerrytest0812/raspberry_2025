```python
# å°å…¥å¿…è¦çš„å¥—ä»¶
import paho.mqtt.client as mqtt
import time
```

```python
# è¨­å®š MQTT å›å‘¼å‡½å¼
def on_connect(client, userdata, flags, reason_code, properties):
    """é€£ç·šæˆåŠŸæ™‚çš„å›å‘¼å‡½å¼"""
    print(f"âœ… é€£ç·šæˆåŠŸï¼ä»£ç¢¼: {reason_code}")
    client.subscribe("test/topic1")
    print("ğŸ“¡ å·²è¨‚é–±ä¸»é¡Œ: test/topic1")

def on_message(client, userdata, msg):
    """æ”¶åˆ°è¨Šæ¯æ™‚çš„å›å‘¼å‡½å¼"""
    print(f"\nğŸ“¨ æ”¶åˆ°è¨Šæ¯ï¼")
    print(f"   ä¸»é¡Œ: {msg.topic}")
    print(f"   å…§å®¹: {msg.payload.decode()}")
    # æ”¶åˆ°è¨Šæ¯å¾Œï¼Œæ–·é–‹é€£ç·šçµæŸæ¸¬è©¦
    client.disconnect()
    print("ğŸ”Œ å·²æ–·é–‹é€£ç·š")

def on_disconnect(client, userdata, disconnect_flags, reason_code, properties):
    """æ–·é–‹é€£ç·šæ™‚çš„å›å‘¼å‡½å¼ï¼ˆVERSION2 APIï¼‰"""
    print(f"ğŸ‘‹ å·²èˆ‡ MQTT Broker æ–·é–‹é€£ç·šï¼ŒåŸå› ä»£ç¢¼: {reason_code}")
```

```python
# å»ºç«‹ MQTT å®¢æˆ¶ç«¯ï¼ˆä½¿ç”¨æœ€æ–°çš„å›èª¿ API ç‰ˆæœ¬ï¼‰
client = mqtt.Client(mqtt.CallbackAPIVersion.VERSION2)
client.on_connect = on_connect
client.on_message = on_message
client.on_disconnect = on_disconnect

print("ğŸ”§ MQTT å®¢æˆ¶ç«¯å·²å»ºç«‹")
```

```python
# é€£ç·šåˆ° MQTT Brokerï¼ˆlocalhost ä»£è¡¨é€£æœ¬æ©Ÿï¼‰
print("ğŸ”— æ­£åœ¨é€£ç·šåˆ° MQTT Broker...")
client.connect("localhost", 1883, 60)

# å•Ÿå‹•ç›£è½è¿´åœˆ
client.loop_start()
print("â³ ç­‰å¾…é€£ç·šç©©å®š...")
time.sleep(1)  # ç­‰å¾…é€£ç·šç©©å®š
```

```python
# ç™¼é€æ¸¬è©¦è¨Šæ¯
print("ğŸ“¤ æ­£åœ¨ç™¼é€æ¸¬è©¦è¨Šæ¯...")
client.publish("test/topic1", "Hello RPi 5, MQTT is working!")
print("âœ… è¨Šæ¯å·²ç™¼é€")
```
